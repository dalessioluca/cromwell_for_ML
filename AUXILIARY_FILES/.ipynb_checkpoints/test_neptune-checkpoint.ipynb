### Set the NEPTUNE_API_TOKEN:
To find your NEPTUNE_API_TOKEN:
1. log into the Neptune website (*https://neptune.ai/*)
2. follow this instruction: https://docs.neptune.ai/security-and-privacy/api-tokens/how-to-find-and-set-neptune-api-token.html

to your *~/.bashrc* or *~/.bash_profile* file

### Minimal test of Neptune logging functionality

```python
import neptune

NEPTUNE_PROJECT = "dalessioluca/sandbox"
neptune.set_project(NEPTUNE_PROJECT)
exp = neptune.create_experiment()

x = 1.0
for i in range(100):
    exp.log_metric("x", x)
    x *= 0.8
exp.stop()
```
